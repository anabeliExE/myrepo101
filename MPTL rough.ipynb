```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(0,10)
```

```python
x
```

```python
y=x ** 2
```

```python
y
```

```python
plt.plot(x,y, 'r-')
plt.xlabel('first 10 whole numbers')
plt.xlabel('squares of the nos')
plt.title('Square')
```

```python
plt.subplot(1,2,1)#firstplot
plt.plot(x,y)
plt.subplot(1,2,2)#second plot
plt.plot(y,x,'r')
```

```python
z={-3,56,8,41,55,46,16,6,-12,20}
```

```python
plt.subplot(1,2,1)#firstplot
plt.plot(x,y)
plt.subplot(1,2,2)#second plot
plt.plot(y,x,'r')
plt.subplot(1,2,3)#second plot
plt.plot(x,z,'r')
```

```python
#we cannot make more than 3 plots in a single subplot
```

```python
#OOM
```

```python
fig=plt.figure()#creating a figure object
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('numbers')
axes.set_ylabel('squares')
```

```python
fig=plt.figure()#creating a figure object
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.52,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
#SUBPLOTS
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
#axes.plot(x,y)
axes[0].plot(x,y)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('rough 1')
axes[1].plot(y,x, 'r')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('rough 2')
```

```python
fig,axes=plt.subplots(nrows=2, ncols=1, figsize=(6,4))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```
